```python
import pandas as pd
```

```python
# import matches.csv file and perform data understanding

matches = pd.read_csv('matches.csv')
matches.head()
```

```python
matches.shape
```

```python
matches.info()
```

```python
# identify total unique values

matches.nunique()
```

```python
# identify duplicate values

matches.duplicated().sum()
```

```python
# total null values

matches.isnull().sum()
```

```python
# Replace missing values for basic columns with appropriate values
matches['city'].fillna('Unknown', inplace=True)  # Replace missing city with 'Unknown'
matches['winner'].fillna('No Result', inplace=True)  # Replace missing winner with 'No Result'
matches['player_of_match'].fillna('None', inplace=True)  # Replace missing player_of_match with 'None'
matches['umpire1'].fillna('Unknown', inplace=True)
matches['umpire2'].fillna('Unknown', inplace=True)
```

```python
matches.isnull().sum()
```

```python
matches.describe()
```

```python
# find total number of matches played in each IPL season and plot graph for the same

matches_per_season = matches['season'].value_counts().sort_index()
```

```python
import matplotlib.pyplot as plt

# Plotting
plt.figure(figsize=(10, 6))
matches_per_season.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Total Matches Played Per IPL Season', fontsize=16)
plt.xlabel('Season', fontsize=12)
plt.ylabel('Number of Matches', fontsize=12)
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# team which won by maximum run and print the detail of match

max_win_by_runs = matches.loc[matches['win_by_runs'].idxmax()]

# Extract relevant details
max_run_details = {
    'Team': max_win_by_runs['winner'],
    'Runs': max_win_by_runs['win_by_runs'],
    'Opponent': max_win_by_runs['team2'] if max_win_by_runs['team1'] == max_win_by_runs['winner'] else max_win_by_runs['team1'],
    'Venue': max_win_by_runs['venue'],
    'Season': max_win_by_runs['season'],
    'Date': max_win_by_runs['date'],
}

max_run_details
```

```python
# find out team which won by maximum wickets


max_win_by_wickets = matches.loc[matches['win_by_wickets'].idxmax()]

# Extract relevant details
max_wicket_details = {
    'Team': max_win_by_wickets['winner'],
    'Wickets': max_win_by_wickets['win_by_wickets'],
    'Opponent': max_win_by_wickets['team2'] if max_win_by_wickets['team1'] == max_win_by_wickets['winner'] else max_win_by_wickets['team1'],
    'Venue': max_win_by_wickets['venue'],
    'Season': max_win_by_wickets['season'],
    'Date': max_win_by_wickets['date'],
}

max_wicket_details
```

```python
# which team is winner in most of the matches plot graph for first 5 winning teams

# Extract winners from the first five matches
first_five_matches = matches.head(5)
first_five_winners = first_five_matches['winner'].value_counts()

# Plotting
plt.figure(figsize=(8, 5))
first_five_winners.plot(kind='bar', color='lightcoral', edgecolor='black')
plt.title('Winners in the First 5 IPL Matches', fontsize=16)
plt.xlabel('Teams', fontsize=12)
plt.ylabel('Number of Wins', fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

first_five_winners
```

```python
# find out chances of winning percentage of teams batting second


# Calculate the total matches and wins for teams batting second
matches_batting_second = matches[matches['win_by_wickets'] > 0]  # Matches won by wickets indicate batting second
total_batting_second_matches = len(matches)
wins_batting_second = len(matches_batting_second)

# Calculate winning percentage for teams batting second
batting_second_win_percentage = (wins_batting_second / total_batting_second_matches) * 100

# Result
batting_second_win_percentage
```

```python
# who were the winners and runner up in each season

# Identifying winners and runner-ups for each season
final_matches = matches.sort_values('date').groupby('season').tail(1)  # Get the last match of each season
winners_and_runners_up = final_matches[['season', 'team1', 'team2', 'winner']]

# Adding runner-up information (team1 or team2 that did not win)
winners_and_runners_up['runner_up'] = winners_and_runners_up.apply(
    lambda x: x['team1'] if x['team2'] == x['winner'] else x['team2'], axis=1
)

winners_and_runners_up.reset_index(drop=True, inplace=True)
winners_and_runners_up
```

```python

# Q1: Analyze toss decisions and create a pie chart
toss_decisions = matches['toss_decision'].value_counts()

# Plotting the pie chart
plt.figure(figsize=(6, 6))
toss_decisions.plot.pie(autopct='%1.1f%%', labels=toss_decisions.index, startangle=90, colors=['#ff9999','#66b3ff'])
plt.title("Analysis of Toss Decisions")
plt.ylabel('')  # Remove y-axis label for better visualization
plt.show()

toss_decisions
```

```python
# Q2: Display data of top players of match winners
top_players = matches['player_of_match'].value_counts().head(10)

# Prepare a bar chart for visualization
plt.figure(figsize=(10, 6))
top_players.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title("Top 10 Players of Match Winners")
plt.xlabel("Player")
plt.ylabel("Number of Awards")
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

top_players
```

```python
# Q3: Count total number of matches played at each venue
venue_counts = matches['venue'].value_counts()

# Plotting the data for venues with the highest number of matches
plt.figure(figsize=(12, 6))
venue_counts.head(10).plot(kind='bar', color='lightgreen', edgecolor='black')
plt.title("Top 10 Venues with Highest Number of Matches")
plt.xlabel("Venue")
plt.ylabel("Number of Matches")
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

venue_counts.head(10)
```
